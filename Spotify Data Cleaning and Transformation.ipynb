```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spotify_history.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Using Forward fill as the values are similar to fill null values 
df['reason_start'].fillna(method='ffill', inplace=True)
df['reason_end'].fillna(method='ffill', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Convert 'ts' to datetime
df['ts'] = pd.to_datetime(df['ts'])
```

```python
# Check for duplicate rows
print(df.duplicated().sum())
```

```python
# Removing duplicates
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
# Extract year, month, day
df['year'] = df['ts'].dt.year
df['month'] = df['ts'].dt.month_name()
df['day'] = df['ts'].dt.day_name()
```

```python
# Convert ms_played from milliseconds to seconds
df['ms_played'] = df['ms_played'] / 1000
```

```python
df.drop('ms_played_sec', axis=1, inplace=True)
```

```python
df.to_csv("spotify_cleaned_trasnformed.csv" , index=False)
```
